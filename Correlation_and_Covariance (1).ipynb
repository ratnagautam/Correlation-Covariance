#Correlation and Covariance


1.Define Covariance and explain how it differs from Correlation in terms of scale and interpretation.

    -Covariance measures how two variables move together. It indicates whether increases in one variable are associated with increases or decreases in another.

    Difference between Covariance and Correlation:

    a)Scale: Covariance depends on the units of measurement, while correlation is unit-free.

    b)Interpretation: Covariance only shows the direction of the relationship, whereas correlation shows both direction and strength of the relationship (between –1 and +1).

2.What does a positive, negative, and zero covariance indicate about the relationship between two variables?

    -a)Positive covariance: Both variables tend to increase or decrease together.

    b)Negative covariance: One variable increases while the other decreases.

    c)Zero covariance: No linear relationship exists between the variables.

3.Discuss the limitations of covariance as a measure of relationship between two variables. Why is correlation
preferred in many cases?

    -Limitations of covariance:

    a)It is affected by the scale of measurement.
    b)The value is difficult to interpret because it has no fixed range.
    c)Cannot compare relationships across different datasets.

    -Why correlation is preferred:

    a)It is standardized and lies between –1 and +1.
    b)Easy to interpret and compare.
    c)Clearly indicates both strength and direction of the relationship.

4. Explain the difference between Pearson’s correlation coefficient and Spearman’s rank correlation
coefficient. When would you prefer to use Spearman’s correlation?

    -Pearson’s correlation: Measures linear relationships using actual values; assumes normal distribution.

    Spearman’s correlation: Measures monotonic relationships using ranked data; does not require normal distribution.

    -Spearman’s correlation is preferred when:

    a)Data is ordinal or ranked.
    b)Relationship is non-linear but monotonic.
    c)Data contains outliers.

5.If the correlation coefficient between two variables X and Y is 0.85, interpret this value in context. Can you
infer causation from this value? Why or why not?

    -A correlation coefficient of 0.85 indicates a strong positive linear relationship between X and Y.

    Causation cannot be inferred because correlation only shows association, not cause-and-effect. Other variables or external factors may influence both X and Y.
